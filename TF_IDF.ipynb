Perform preprocessing steps on the ”Heading” field of the dataset i.e. lowercase, stopword removal,
stemming

```python
# Preprocessing function
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))
    # Tokenization and stemming
    stemmer = SnowballStemmer("english")
    tokens = [stemmer.stem(word) for word in text.split()]
    # Remove stopwords
    stop_words = set(stopwords.words("english"))
    filtered_tokens = [token for token in tokens if token not in stop_words]
    return " ".join(filtered_tokens)
```

```python
# Apply preprocessing to the 'Heading' field
df['Heading'] = df['Heading'].apply(preprocess_text)
```

TF-IDF from scratch

```python
#import library
import numpy as np
```

```python
# Calculate term frequency (TF)
def calculate_tf(document):
    words = document.split()
    tf_dict = {}
    for word in words:
        tf_dict[word] = tf_dict.get(word, 0) + 1
    total_words = len(words)
    for word, count in tf_dict.items():
        tf_dict[word] = count / total_words
    return tf_dict
```

```python
# Calculate inverse document frequency (IDF)
def calculate_idf(documents):
    idf_dict = {}
    total_documents = len(documents)
    for document in documents:
        words = set(document.split())
        for word in words:
            idf_dict[word] = idf_dict.get(word, 0) + 1
    for word, count in idf_dict.items():
        idf = np.log(total_documents / (count + 1))  # Adding 1 to avoid division by zero
        idf_dict[word] = idf
    return idf_dict
```

```python
# Calculate TF-IDF matrix
def calculate_tfidf(documents):
    tfidf_matrix = []
    idf_dict = calculate_idf(documents)
    for document in documents:
        tf_dict = calculate_tf(document)
        tfidf_vector = [tf_dict.get(word, 0) * idf_dict[word] for word in idf_dict.keys()]
        tfidf_matrix.append(tfidf_vector)
    return np.array(tfidf_matrix)
```

```python
# Create a list of preprocessed documents
documents = df['Heading'].tolist()
print(documents)
```

```python
# Calculate TF-IDF matrix
tfidf_matrix = calculate_tfidf(documents)
print(tfidf_matrix)
```

```python
# Display the shape of the TF-IDF matrix
print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```
