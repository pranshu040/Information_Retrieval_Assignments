```python
#importing packages
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
#creating a set of stopwords
stop_words = set(stopwords.words('english'))

#tokenization
tokens = word_tokenize(Articles)

#printing the list of 'new_words'
new_words = [word for word in tokens if word.lower() not in stop_words]
print(new_words)
```

```python
#importing packages
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw')
```

```python
from nltk.stem import PorterStemmer
porter_stem = PorterStemmer()

#printing the list of 'stem_words'
stem_words = [porter_stem.stem(word) for word in tokens]
print(stem_words)
```

```python
invert_index = {}
for _, row in train_d1.iterrows():
    article_ids, article_texts = row['id'], row['Article']

    # Tokenize the article text
    tokens = word_tokenize(article_texts)

    # Remove stopwords and perform stemming
    new_words = [porter_stem.stem(word.lower()) for word in tokens if word.lower() not in stop_words]

    # Create the inverted index
    for word in new_words:
        if word not in invert_index:
            invert_index[word] = [article_ids]
        else:
            invert_index[word].append(article_ids)

words_to_display = ['pakistan', 'pm', 'imran', 'khan', 'take', 'swipe']
for word in words_to_display:
    print(f"Word: {word}, Inverted Index: {invert_index.get(word, [])}")



```
