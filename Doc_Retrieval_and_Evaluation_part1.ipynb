```python
#Extracting the document id and text using pandas dataframe

import os
from bs4 import BeautifulSoup

years = []
subfolders = []
doc_numbers = []
doc_texts = []

base_path = "/content/drive/MyDrive/IR/FIRE_Dataset_EN_2010/TELEGRAPH_UTF8"
```

```python
# Loop through each year folder in the base directory
for year_folder in os.listdir(base_path):
  year = year_folder[:4] #Extract year from folder name
  year_folder_path = os.path.join(base_path, year_folder)

  #Loop through each subfolder within year folder
  for subfolder in os.listdir(year_folder_path):
    subfolder_path = os.path.join(year_folder_path, subfolder)

    #Loop through XML file within subfolder
    for xml_file in os.listdir(subfolder_path):
      xml_file_path = os.path.join(subfolder_path, xml_file)

      #Parse the xml file within each subfolder
      with open(xml_file_path, 'r', encoding='utf-8') as file:
        xml_content = file.read()
        soup = BeautifulSoup(xml_content, 'xml')

        #Extract doc no and text
        docno = soup.find('DOCNO').text.strip()
        text = soup.find('TEXT').text.strip()

        #Append extracted data to lists
        years.append(year)
        subfolders.append(subfolder)
        doc_numbers.append(docno)
        doc_texts.append(text)
```

```python
data = {'Year': years, 'Subfolder': subfolders, 'Doc_no': doc_numbers, 'Text': doc_texts}
```

```python
all_docs_df = pd.DataFrame(data)
```

```python
#remove punctutation
def remove_punct(text):
    return ("".join([ch for ch in text if ch not in st.punctuation]))
```

```python
#tokenize the text
def tokenize(text):
    text = re.split('\s+' ,text)
    return [x.lower() for x in text]
```

```python
#remove smallwords
def remove_small_words(text):
    return [x for x in text if len(x) > 3 ]
```

```python
#remove stopwords
def remove_stopwords(text):
    return [word for word in text if word not in nltk.corpus.stopwords.words('english')]
```

```python
#stemming
def stemming(text):
    ps = PorterStemmer()
    return [ps.stem(word) for word in text]
```

```python
#lemmatization
def lemmatize(text):
    word_net = WordNetLemmatizer()
    return [word_net.lemmatize(word) for word in text]
```

```python
#return tokens
def return_sentences(tokens):
    return " ".join([word for word in tokens])
```

```python
#Creating a dataframe to store the query vector and titles

data = {'qid':all_nums, 'query': all_titles}
query_df = pd.DataFrame(data)
```

```python
query_df.head()
```

```python
query_df['query'] = query_df['query'].apply(lambda x: remove_punct(x))
query_df['query'] = query_df['query'].apply(lambda msg : tokenize(msg))
query_df['query'] = query_df['query'].apply(lambda x : remove_small_words(x))
query_df['query'] = query_df['query'].apply(lambda x : remove_stopwords(x))
query_df['query'] = query_df['query'].apply(lambda x : lemmatize(x))
query_df['query'] = query_df['query'].apply(lambda x : return_sentences(x))
```

```python
query_df.head()
```

```python
query_df.shape
```

```python
tfidf = pt.BatchRetrieve(index, controls={"wmodel": "TF_IDF"})
# Remove single quotes from queries
query_df['query'] = query_df['query'].str.replace("'", "")

# Use the processed queries to retrieve documents
results = tfidf.transform(query_df)
```

```python
results.head()
```

```python
# Read the qrels file and create a DataFrame 'q_df'
path_to_qrels = "/content/drive/MyDrive/IR/FIRE_Dataset_EN_2010/en.qrels.76-125.2010.txt"
q_df = pt.io.read_qrels(path_to_qrels)
```

```python
# Evaluate using precision at 10 documents and recall at 10 documents
eval_metrics = pt.Experiment(
    [tfidf],
    results,
    q_df,
    eval_metrics=["P_10", "recall_10"])
```

```python
eval_metrics
```

```python
f1 = 2 * (eval_metrics['P_10'] * eval_metrics['recall_10']) / eval_metrics['P_10'] + eval_metrics['recall_10']

print("F1 Score:", f1)
```
